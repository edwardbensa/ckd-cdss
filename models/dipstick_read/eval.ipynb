```python
import pandas as pd
from src.config import MODELS_DIR
df1 = pd.read_csv(MODELS_DIR / "dipstick_read/plain_03/results.csv")
df2 = pd.read_csv(MODELS_DIR / "dipstick_read/simple_03/results.csv")
df3 = pd.read_csv(MODELS_DIR / "dipstick_read/granular_01/results.csv")
df4 = pd.read_csv(MODELS_DIR / "dipstick_read/rotation_04/results.csv")
```

```python
cols = ["epoch", "time", "metrics/precision(B)", "metrics/recall(B)", "metrics/mAP50(B)", "metrics/mAP50-95(B)"]
add_cols = ["train/box_loss", "train/cls_loss", "train/dfl_loss", "val/box_loss", "val/cls_loss", "val/dfl_loss"]
cols = cols + add_cols
```

```python
final = pd.DataFrame(cols, columns=["metric"])
for cn, df in zip(["plain", "simple", "granular", "rotation"], [df1, df2, df3, df4]):
    final[cn] = df.iloc[-1:][cols].transpose().values
final
```
